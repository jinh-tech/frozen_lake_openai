```python
import gym
import numpy as np
import matplotlib.pyplot as plt
```

```python
env = gym.make("FrozenLake-v0")
```

```python
q = np.zeros(shape=(env.observation_space.n,env.action_space.n))
learning_rate = 1.0
discount = 0.95
epsilon = 0.1
num_episodes = 100000
reward_array = []
lam = 0.99
```

```python
for i in range(0,num_episodes):
    state = env.reset()
    trace = np.zeros(shape=(env.observation_space.n,env.action_space.n))
    action = np.random.randint(env.action_space.n)
    reward_episode = 0.0
    for j in range(0,10000):
        new_state,reward,done,_ = env.step(action)
        if(np.random.uniform()>epsilon):
            new_action = np.argmax(q[new_state,:])
        else:
            new_action = np.random.randint(env.action_space.n)
        
        trace[state,action] += 1.0
        delta = reward + discount*q[new_state,new_action] - q[state,action]
        
        for s in range(0,env.observation_space.n):
            for a in range(0,env.action_space.n):
                q[s,a] = q[s,a] + learning_rate*trace[s,a]*delta
                trace[s,a] = discount*lam*trace[s,a]
        
        reward_episode += reward
        state = new_state
        action = new_action
        
        if done:
            reward_array.append(reward_episode)
            break            
```

```python
plot_reward = []
plot_reward.append(0)
for i in range(0,100):
    plot_reward[-1] += reward_array[i]
for i in range(0,len(reward_array)-100):
    plot_reward.append(plot_reward[-1]-reward_array[i]+reward_array[i+100])

for i in range(0,len(plot_reward)):
    plot_reward[i]/=100.0


    
for i in range(1,len(reward_array)):
    reward_array[i] += reward_array[i-1] 
for i in range(1,len(reward_array)):
    reward_array[i] /= (i+1)
    
plt.figure()
plt.plot(reward_array)
plt.show()
plt.figure()
plt.plot(plot_reward)
plt.show()
```
